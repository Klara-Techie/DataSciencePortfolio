```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2)
```

```python
kmeans.fit(clg.drop('Private', axis=1))
```

```python
kmeans.cluster_centers_
```

```python
def privateOrPublic(col):
    if col == 'Yes':
        return 1
    else:
        return 0
```

```python
clg['Cluster'] = clg['Private'].apply(privateOrPublic)
```

```python
clg.head()
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(clg['Cluster'], kmeans.labels_))
```

```python
print(classification_report(clg['Cluster'], kmeans.labels_))
```
